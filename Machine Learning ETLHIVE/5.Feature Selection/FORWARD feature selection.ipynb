# Forward Feature selection

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

## Step 1:- Reading of the dataset

```python
import pandas as pd 
df = pd.read_csv("Cars93.csv", na_values=["", "NA"], keep_default_na=False)
df.head()
```

### Step 2 :- Perform Basic Data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m[m>0]
```

```python
df.duplicated().sum()
```

## Step 3:- Seperate X and Y(Weight)
drop id columns

```python
X = df.drop(columns=["id", "Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

## Step 4:- Feature Selection Pipeline

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
# Feature selection pipeline only
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer 
```

```python
num_pipe1 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe1 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder()
)
```

```python
pre1 = ColumnTransformer([
    ("num", num_pipe1, con),
    ("cat", cat_pipe1, cat)
]).set_output(transform="pandas")
```

```python
X_pre1 = pre1.fit_transform(X)
X_pre1.head()
```

```python
pre1
```

# Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
sel = SequentialFeatureSelector(
    estimator=LinearRegression(), direction="forward", n_features_to_select="auto"
)
sel.fit(X_pre1, Y)
```

```python
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel.get_feature_names_out())
```

```python
sel_cols[0]
```

```python
sel_cols[0].split("__")[1]
```

```python
imp_cols = [col.split("__")[1] for col in sel_cols]
imp_cols
```

```python
X_sel = X[imp_cols]
```

```python
X_sel.head()
```

## Step 5:- Create a final preprocessing pipeline

```python
cat_sel = list(X_sel.columns[X_sel.dtypes == "object"])
con_sel = list(X_sel.columns[X_sel.dtypes != "object"])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)
)
```

```python
pre2 = ColumnTransformer([
    ("num", num_pipe2, con_sel),
    ("cat", cat_pipe2, cat_sel)
]).set_output(transform="pandas")
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
pre2
```

## Step 6:- Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    X_sel_pre, Y, test_size=0.2, random_state=21
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Step 7:- Build the model

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
coef = pd.Series(model.coef_.flatten(), index=xtrain.columns)
coef
```

```python
top10.plot(kind='bar')
```

```python
bottom10 = coef.sort_values().head(10)
bottom10
```

```python
bottom10.plot(kind="bar")
```

## Step 8 :- Model Evaluation

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(ypred, y)
    rmse = mse**(1/2)
    mae = mean_absolute_error(ypred, y)
    mape = mean_absolute_percentage_error(ypred, y)
    r2 = r2_score(y, ypred)
    print(f"MSE : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 score : {r2:.4f}")
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```

## From above both train and test have the r2 score > 0.8 and r2 score improved after using the feature selection method
